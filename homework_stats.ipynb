# Задание 2. Работа с выбросами¶

В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.


```python
# нормальный пульс 30-40, предполагаю что пульс больше чем в два раза от нормального состояния является выбросом
badpulse = horse[horse.pulse < 80]
badpulse
```

```python
# Pulse до выброса
print(horse.pulse.max())
print(horse.pulse.min())
print(horse.pulse.max() - horse.pulse.min(), 'размах')
print(horse.pulse.mode()[0],'мода') 
print(horse.pulse.median(),'медиана')
print(horse.pulse.var(),'дисперсия')
```

```python
# pulse после выброса
print(badpulse.pulse.max())
print(badpulse.pulse.min())
print(badpulse.pulse.max() - badpulse.pulse.min(), 'размах')
print(badpulse.pulse.mode()[0],'мода') 
print(badpulse.pulse.median(),'медиана')
print(badpulse.pulse.var(),'дисперсия')
```

```python
#выбросы через межквантильный размах
q1 = horse.pulse.quantile(0.25)
q3 = horse.pulse.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
remove_outliera = horse[horse.pulse.between(lower_bound,upper_bound,inclusive=True)]
remove_outliera
```

```python
pd.concat([horse,remove_outliera]).drop_duplicates(keep=False)# выбросы
```

## вывод:
работая с выбросами по столбцу "Pulse" я увидел,что большинство данных отсутсвуют и не попадают в указанный диапозон.
Предполагаю,что причина выбросов является не качественный эксперимент, т.е. экспериментальные ошибки и ошибка ввода данных.Слишком велико значение размаха, из-за этого данные искажаются.


# Задание 3. Работа с пропусками

Рассчитать количество выбросов для всех выбранных столбцов. Принять и обосновать решение о методе работы с пропусками по каждому столбцу, сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
for col in horse.columns:
    pot_missing = horse[col].isnull().mean()
    print(f'{col} - {pot_missing:.1%}')
```

```python
horse['age'].fillna('no info').isna().sum()
```

```python
horse2 = data[['surgery','age','pulse','abdominal_distension','abdomen','outcome',
              'surgical lesion','type of lesion']]
```

```python
#заполнение пропусков в столбце abdomen на значение mode(использую моду как часто используемое значение в столбце)
horse2.abdomen.fillna(horse.abdomen.median() ,inplace=True)
```

```python
#заполнение пропусков в столбце abdominal distension на значение mode
horse2.abdominal_distension.fillna(horse.abdominal_distension.median() ,inplace=True)
```

#### заполнение пропусков с помощью группировки(использую медиану для заполнения,так как результат зависел от хирургического вмешательства и от значения пульса)


```python
horse2.pulse.fillna(horse2.groupby(['outcome','surgery']).pulse.transform('median'),inplace=True)
horse2.groupby(['outcome','surgery']).pulse.median()
```
